```python
def calculate_average():
    print("--- Student Average Marks Calculator ---")

    # Get the number of subjects
    try:
        num_subjects = int(input("Enter the number of subjects: "))
        if num_subjects <= 0:
            print("Please enter a number greater than 0.")
            return
    except ValueError:
        print("Invalid input. Please enter a whole number.")
        return

    marks = []

    # Loop to get marks for each subject
    for i in range(1, num_subjects + 1):
        while True:
            try:
                mark = float(input(f"Enter marks for subject {i}: "))
                if 0 <= mark <= 100:  # Assuming marks are out of 100
                    marks.append(mark)
                    break
                else:
                    print("Please enter a valid mark between 0 and 100.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    # Calculate total and average
    total = sum(marks)
    average = total / num_subjects
55
    # Print Results
    print("\n--- Results ---")
    print(f"Total Marks Scored: {total} out of {num_subjects * 100}")
    print(f"Average Percentage: {average:.2f}%")

# Run the program
calculate_average()
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# Paparan Tajuk dan CSS
display(HTML("""
<style>
.card {
    background: linear-gradient(135deg, #74ebd5, #9face6);
    padding: 25px;
    border-radius: 20px;
    box-shadow: 0px 4px 15px rgba(0,0,0,0.2);
    width: 500px;
    margin: auto;
    text-align: center;
}

.title {
    font-size: 28px;
    font-weight: bold;
    color: black;
    margin-bottom: 15px;
}

.result {
    padding: 15px;
    border-radius: 10px;
    margin-top: 15px;
    font-size: 18px;
    font-weight: bold;
}
</style>

<div class="card">
    <div class="title">
        🏃 BMI CALCULATOR 🏃
    </div>
</div>
"""))

# Input Berat
berat = widgets.FloatText(
    description='⚖ Berat (kg)',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='350px')
)

# Input Tinggi
tinggi = widgets.FloatText(
    description='📏 Tinggi (cm)',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='350px')
)

# Output
output = widgets.Output()

# Fungsi Kira BMI
def kira_bmi(btn):
    output.clear_output()

    with output:
        try:
            b = berat.value
            t = tinggi.value

            if b <= 0 or t <= 0:
                display(HTML("""
                <div class='result' style='background:#ffebee;color:red;'>
                ❌ Sila masukkan nilai yang sah!
                </div>
                """))
                return

            bmi = b / ((t / 100) ** 2)

            if bmi < 18.5:
                kategori = "Kurang Berat Badan"
                warna = "#FFA726"
                icon = "⚠️"

            elif bmi < 25:
                kategori = "Berat Badan Normal"
                warna = "#66BB6A"
                icon = "✅"

            elif bmi < 30:
                kategori = "Berlebihan Berat Badan"
                warna = "#FFCA28"
                icon = "⚡"

            else:
                kategori = "Obesiti"
                warna = "#EF5350"
                icon = "🚨"

            display(HTML(f"""
            <div class='result'
                 style='background:{warna};color:black;'>
                 {icon}<br>
                 BMI Anda: <b>{bmi:.2f}</b><br>
                 Kategori: <b>{kategori}</b>
            </div>
            """))

        except:
            display(HTML("""
            <div class='result' style='background:#ffebee;color:red;'>
            ❌ Input tidak sah!
            </div>
            """))

# Butang
btn = widgets.Button(
    description='KIRA BMI',
    button_style='success',
    layout=widgets.Layout(width='200px', height='45px')
)

btn.on_click(kira_bmi)

# Jadual BMI
jadual = HTML("""
<div style="
background:black;
padding:15px;
border-radius:10px;
width:500px;
margin:auto;
box-shadow:0px 2px 10px rgba(0,0,0,0.1);
">
<h3>📊 Jadual BMI</h3>
<ul>
<li><b>BMI < 18.5</b> → Kurang Berat Badan</li>
<li><b>18.5 - 24.9</b> → Berat Badan Normal</li>
<li><b>25.0 - 29.9</b> → Berlebihan Berat Badan</li>
<li><b>BMI ≥ 30</b> → Obesiti</li>
</ul>
</div>
""")

# Paparan Interface
display(
    widgets.VBox(
        [
            berat,
            tinggi,
            btn,
            output
        ],
        layout=widgets.Layout(
            align_items='center',
            padding='20px'
        )
    )
)

display(jadual)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML

# Tajuk
display(HTML("<h1 style='color:blue;text-align:center;'>SISTEM PENGIRAAN BMI</h1>"))

# Widget input
berat_input = widgets.FloatText(
    description='Berat (kg):',
    style={'description_width': 'initial'}
)

tinggi_input = widgets.FloatText(
    description='Tinggi (cm):',
    style={'description_width': 'initial'}
)

# Output
output = widgets.Output()

# Fungsi BMI
def kira_bmi(b):
    output.clear_output()

    berat = berat_input.value
    tinggi = tinggi_input.value

    with output:
        if berat <= 0 or tinggi <= 0:
            print("❌ Berat dan tinggi mestilah lebih daripada 0!")
            return

        bmi = berat / ((tinggi / 100) ** 2)

        if bmi < 18.5:
            kategori = "Kurang Berat Badan"
            warna = "orange"
        elif bmi < 25:
            kategori = "Berat Badan Normal"
            warna = "green"
        elif bmi < 30:
            kategori = "Berlebihan Berat Badan"
            warna = "gold"
        else:
            kategori = "Obesiti"
            warna = "red"

        display(HTML(
            f"""
            <h2 style='color:{warna};'>
            BMI Anda: {bmi:.2f}<br>
            Kategori: {kategori}
            </h2>
            """
        ))

# Butang
btn_kira = widgets.Button(
    description='KIRA BMI',
    button_style='success'
)

btn_kira.on_click(kira_bmi)

# Jadual BMI
info = widgets.HTML(
"""
<h3>Jadual BMI</h3>
<ul>
<li><b>BMI < 18.5</b> : Kurang Berat Badan</li>
<li><b>18.5 - 24.9</b> : Berat Badan Normal</li>
<li><b>25.0 - 29.9</b> : Berlebihan Berat Badan</li>
<li><b>BMI ≥ 30</b> : Obesiti</li>
</ul>
"""
)

# Paparan
display(
    widgets.VBox([
        berat_input,
        tinggi_input,
        btn_kira,
        output,
        info
    ])
)
```
